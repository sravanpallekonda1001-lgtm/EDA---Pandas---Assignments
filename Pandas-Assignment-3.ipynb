## Difference between groupby and pivot_table
-Groupby is useful for grouping data by one or more columns and calculating aggregate functions. Pivot tab useful fro summarizing and aggregating data in a tabular format output: The output of the groupby and pivot table is different .Groupby returns a pandas series or DataFrame with the affregated data while p returns pandas DataFrame with a tabular format. Aggregation: Group by and pivot table handle aggregation differently.Group by performs aggregation on a column or set of columns while pivot performs aggregation on theintersection of rows and columns.

```python
df = pd.read_csv("C:\\Users\\srava_kv\\Downloads\\sales_dataset_2000_rows.csv")
```

```python
df
```

Find the total quantity sold in each Region.

```python
result=df.groupby("Region")["Quantity"].sum()
print(result)
```

Find the average Unit_Price for each Product_Category.

```python
df.groupby("Product_Category")["Unit_Price"].mean()
```

Calculate the total sales quantity by City.

```python
df.groupby("City")["Quantity"].sum()
```

Find the maximum discount given in each Region.

```python
df.groupby("Region")["Discount"].max()
```

Find the total quantity sold by Region and Product_Category.

```python
df.groupby(["Product_Category","Region"])["Quantity"].sum()
```

Find the average Unit_Price by Region and Payment_Mode.

```python
df.groupby(["Region","Payment_Mode"])["Unit_Price"].mean()
```

Find the total quantity sold by Gender and Region.

```python
gender_counts = df.groupby("Region")["Gender"].value_counts().reset_index(name="Count")
print(gender_counts)
```

```python
df.columns = df.columns.str.strip()
```

Find the minimum and maximum Unit_Price for each Product_Category using aggregation.

```python
df.groupby("Product_Category")["Unit_Price"].agg(["min","max"])
```

Find the total and average quantity sold by each Sales_Person using aggregation.

```python
result=df.groupby("Sales_Person")["Quantity"].agg(Total_Quantity="sum",Total_average="mean")
result
```

Find the minimum, maximum, and average discount for each Region using aggregation

```python
df.groupby("Region")["Discount"].agg(The_minimum="min",The_Maximum="max",Discount_="mean")
```
