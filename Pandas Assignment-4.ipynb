## pandas Merge

```python
import pandas as pd
```

```python
df_students = pd.DataFrame({
    "Student_ID": [1, 2, 3, 4],
    "Name": ["Amit", "Ravi", "Sita", "Neha"]
})

df_marks = pd.DataFrame({
    "Student_ID": [2, 3, 4, 5],
    "Marks": [85, 90, 75, 60]
})

df_courses = pd.DataFrame({
    "Student_ID": [1, 2, 3],
    "Course": ["Math", "Science", "English"]
})
```

1. Perform an inner merge between df_students and df_marks ( INNER JOIN) and display the result.

```python
pd.merge(df_students,df_marks,on="Student_ID",how="inner")
```

2. Perform a left merge between df_students and df_marks(LEFT JOIN) and display the result.

```python
df=pd.merge(df_students,df_marks,on="Student_ID",how="left")
df
```

3. Perform a right merge between df_students and df_marks(RIGHT JOIN) and display the result.

```python
pd.merge(df_students, df_marks, on="Student_ID", how="right")
```

4. Perform a full outer merge between df_students and df_marks( OUTER JOIN) and display the result.

```python
pd.merge(df_students, df_marks, on="Student_ID",how="outer")
```

```python
pd.merge(df_students, df_courses, how="cross")
```

6. Perform an inner merge between df_students and df_courses, then merge the result with df_marks (INNER JOIN ) and display the result.

```python
inner_join=pd.merge(df_students, df_courses, on="Student_ID", how="inner")
```

```python
result=pd.merge(inner_join,df_marks,how="inner")
result
```

7. Perform a left merge between df_students and df_marks, then display students who have no marks (LEFT JOIN + CONDITION) and display the result.

```python
marks=pd.merge(df_students, df_marks, on="Student_ID", how="left")
marks
```

```python

marks_no = marks[marks["Marks"].isnull()]

print(marks_no)
```

8.Perform a right merge between df_students and df_marks, then display records where student details are missing (RIGHT JOIN + FILTER)and display the result.


```python
right_merge = pd.merge(df_students, df_marks, on="Student_ID", how="right")
missing_values = right_merge[right_merge["Name"].isnull()]
print(missing_values)
```

9. OUTER JOIN + IDENTIFY
Perform an outer merge between df_students and df_marks, and identify:
only students
only marks
matching records   #Hint: indicator=True

```python
outer_merge = pd.merge(df_students, df_marks, on="Student_ID", how="outer", indicator=True)

print(outer_merge)
```

10. Perform a cross merge between df_students and df_courses, then count the total number of combinations generated (CROSS JOIN )and display the result.

```python
merge=pd.merge(df_students,df_courses,how="cross")
print(len(merge))
```
