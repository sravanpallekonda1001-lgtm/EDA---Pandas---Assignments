# Section 2: Handling Missing Values

### Question 1

```python
data = {
'Name': ['Amit', 'Priya', 'Rahul', 'Sneha'],
'Salary': [50000, None, 45000, None]
}
df=pd.DataFrame(data)
df
```

Identify missing values.

```python
df[df['Salary'].isnull()]
```


Count missing values.

```python
print(df["Salary"].isnull().sum())
```

Fill missing values with mean salary.

```python
missing_salary=df["Salary"].mean()
```

```python
df["Salary"]=df["Salary"].fillna(missing_salary)
```

```python
print(df)
```

## Handling Outliers

```python
data = {
'Sales': [200, 220, 250, 240, 260, 1000]
}
df=pd.DataFrame(data)
df
```

```python
Q1=df["Sales"].quantile(0.25)
Q2=df["Sales"].quantile(0.5)
Q3=df["Sales"].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
lower_boundary=Q1-1.5*IQR
Higher_boundary=Q3+1.5*IQR
```

```python
result=df[(df["Sales"]<lower_boundary)|(df["Sales"]>Higher_boundary)]
```

```python
result
```
