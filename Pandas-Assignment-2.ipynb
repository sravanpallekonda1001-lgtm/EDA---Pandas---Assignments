## Filtering & Logical Thinking

Create a new column called Salary_Per_Experience and analyze which employee is getting paid the most per year of experience.

```python
df["Salary_Per_Experience"] = df["Salary"] / df["Experience_Years"]

top_employee = df.loc[df["Salary_Per_Experience"].idxmax()]
```
